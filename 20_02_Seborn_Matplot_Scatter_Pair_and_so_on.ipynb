```python
# show the value counts of origin column in bar chart
# show the proportion of cars of usa and non usa
df['Origin'].value_counts().plot(kind='bar')
```

```python
# show the proportion of cars of usa and non usa
df['Cylinders'].value_counts().plot(kind='pie')
```

```python
# show the percentage of each type of cars 
df['Type'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
# how is the distribution of horsepower columni
import seaborn as sns
sns.histplot(data=df,x='Horsepower')
```

```python
# check for outliers in length column
sns.boxplot(data=df,x='Length')
```

```python
# check outliers in enginesize column
sns.boxplot(data=df,x='EngineSize')
```

```python
# how is the data distributed in weight column
sns.histplot(data=df,x='Weight')
```

```python
# show the proportions of categories of all categorical columns
from matplotlib.pyplot import show
for i in df.columns:
    if df[i].dtype==object:
        sns.pie(data=df,x=i)
        show()
```

```python
# check for outliers of all numric columns 
from matplotlib.pyplot import show
for i in df.columns:
    if df[i].dtype!=object:
        sns.boxplot(data=df,x=i)
        show()
        
```

```python
# plot the bar chart showing counts of values of all categorical column and histogram showing distribution of all numeric columns
from matplotlib.pyplot import show
for i in df.columns:
    if df[i].dtypes==object:
        df[i].value_counts().plot(kind='bar')
    else:
        sns.histplot(data=df,x=i,kde=True)
        show()
```
