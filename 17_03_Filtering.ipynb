```python
# show the models of type midsize having enginesize more than 3
df[['EngineSize','Type']][(df['Type']=='Midsize')&(df['EngineSize']>3)]
```

```python
# show the models of usa of midsize type having price more than 10 and weight less than 4000
df[['Model','Type','Weight','Price']][(df['Model']=='USA')&(df['Type']=='Midsize')&(df['Weight']<4000)&(df['Price']>10)]
```

```python
''' show the models having price from 20 to 30 and are from usa having weight more than 3300 '''
df[['Model','Price','Weight','Origin']][(df['Price']>20)&(df['Price']<30)&(df['Origin']=='USA')&(df['Weight']>3300)]
```

```python
''' show the models of small or large type having price more than 15
and weight more than 3000
'''
df[['Model','Type','Price','Weight']][(df['Model']=='Small')|(df['Model']=='Large')|(df['Price']>15)|(df['Weight']>3000)]
```

```python
''' show the models of type small or midsize having weight from 3000 to 4000 and price less than 40
'''
df[['Model', 'Type', 'Price', 'Weight']][
    (df['Model'] == 'Small') | 
    (df['Model'] == 'Large') | 
    (df['Price'] > 15) | 
    (df['Weight'] > 3000)
]
```
